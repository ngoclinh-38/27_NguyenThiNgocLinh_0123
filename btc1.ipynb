```python
# bai 1:
class HinhChuNhat:
    def __init__(self):
        self.chieu_dai = 0
        self.chieu_rong = 0
    
    def nhap_du_lieu(self):
        self.chieu_dai = float(input("Nhập chiều dài: "))
        self.chieu_rong = float(input("Nhập chiều rộng: "))
    
    def tinh_chu_vi(self):
        return 2 * (self.chieu_dai + self.chieu_rong)
    
    def tinh_dien_tich(self):
        return self.chieu_dai * self.chieu_rong
    
    def in_thong_tin(self):
        chu_vi = self.tinh_chu_vi()
        dien_tich = self.tinh_dien_tich()
        print(f"Chiều dài: {self.chieu_dai}")
        print(f"Chiều rộng: {self.chieu_rong}")
        print(f"Chu vi: {chu_vi}")
        print(f"Diện tích: {dien_tich}")

hcn = HinhChuNhat()
hcn.nhap_du_lieu()
hcn.in_thong_tin()
```

```python
# bai 2:
class TS:
    def __init__(self):
        self.ho_ten = ""
        self.diem_toan = 0.0
        self.diem_ly = 0.0
        self.diem_hoa = 0.0
    
    def nhap_thong_tin(self):
        while True:
            self.ho_ten = input("Nhập họ tên thí sinh: ")
            if self.ho_ten.strip():
                break
            print("Họ tên không được để trống. Vui lòng nhập lại.")
        
        while True:
            try:
                self.diem_toan = float(input("Nhập điểm Toán: "))
                if 0 <= self.diem_toan <= 10:
                    break
                print("Điểm phải từ 0 đến 10. Vui lòng nhập lại.")
            except ValueError:
                print("Điểm phải là số hợp lệ. Vui lòng nhập lại.")
        
        while True:
            try:
                self.diem_ly = float(input("Nhập điểm Lý: "))
                if 0 <= self.diem_ly <= 10:
                    break
                print("Điểm phải từ 0 đến 10. Vui lòng nhập lại.")
            except ValueError:
                print("Điểm phải là số hợp lệ. Vui lòng nhập lại.")
        
        while True:
            try:
                self.diem_hoa = float(input("Nhập điểm Hoá: "))
                if 0 <= self.diem_hoa <= 10:
                    break
                print("Điểm phải từ 0 đến 10. Vui lòng nhập lại.")
            except ValueError:
                print("Điểm phải là số hợp lệ. Vui lòng nhập lại.")
    
    def in_thong_tin(self):
        print(f"Họ tên: {self.ho_ten}")
        print(f"Điểm Toán: {self.diem_toan}")
        print(f"Điểm Lý: {self.diem_ly}")
        print(f"Điểm Hoá: {self.diem_hoa}")
        print(f"Tổng điểm: {self.tinh_tong_diem()}")
    
    def tinh_tong_diem(self):
        return self.diem_toan + self.diem_ly + self.diem_hoa

n = int(input("Nhập số lượng thí sinh: "))
danh_sach_ts = []

for i in range(n):
    ts = TS()
    print(f"Nhập thông tin thí sinh thứ {i+1}:")
    ts.nhap_thong_tin()
    danh_sach_ts.append(ts)

trung_tuyen = [ts for ts in danh_sach_ts if ts.tinh_tong_diem() >= 20]

trung_tuyen.sort(key=lambda ts: ts.tinh_tong_diem(), reverse=True)

print("\nDanh sách thí sinh trúng tuyển (theo thứ tự điểm từ cao xuống thấp):")
if not trung_tuyen:
    print("Không có thí sinh nào trúng tuyển.")
else:
    for ts in trung_tuyen:
        ts.in_thong_tin()
        print("---")
```

```python
# bai 4:
class Stack:
    def __init__(self, capacity=10):
        self.capacity = capacity
        self.items = []
    
    def __del__(self):
        self.items = None
    
    def push(self, item):
        if not self.isFull():
            self.items.append(item)
            return True
        return False
    
    def pop(self):
        if not self.isEmpty():
            return self.items.pop()
        return None
    
    def isEmpty(self):
        return len(self.items) == 0
    
    def isFull(self):
        return len(self.items) >= self.capacity
```

```python
# bai 8:
class Date:
    def __init__(self, day=1, month=1, year=2000):
        self.day = day
        self.month = month
        self.year = year
    
    def display(self):
        print(f"{self.day}/{self.month}/{self.year}")
    
    def is_leap_year(self):
        return self.year % 4 == 0 and (self.year % 100 != 0 or self.year % 400 == 0)
    
    def days_in_month(self):
        days = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        if self.month == 2 and self.is_leap_year():
            return 29
        return days[self.month]
    
    def next(self):
        self.day += 1
        if self.day > self.days_in_month():
            self.day = 1
            self.month += 1
            if self.month > 12:
                self.month = 1
                self.year += 1

class Employee:
    def __init__(self, ho_ten="", ngay_sinh=None, ngay_vao_cong_ty=None):
        self.ho_ten = ho_ten
        self.ngay_sinh = ngay_sinh if ngay_sinh else Date()
        self.ngay_vao_cong_ty = ngay_vao_cong_ty if ngay_vao_cong_ty else Date()
    
    def nhap_thong_tin(self):
        self.ho_ten = input("Nhập họ tên: ")
        while True:
            try:
                print("Nhập ngày sinh:")
                day = int(input("Ngày (1-31): "))
                month = int(input("Tháng (1-12): "))
                year = int(input("Năm: "))
                if 1 <= day <= 31 and 1 <= month <= 12 and year > 0:
                    self.ngay_sinh = Date(day, month, year)
                    break
                print("Ngày hoặc tháng không hợp lệ. Vui lòng nhập lại.")
            except ValueError:
                print("Vui lòng nhập số nguyên hợp lệ.")
        
        while True:
            try:
                print("Nhập ngày vào công ty:")
                day = int(input("Ngày (1-31): "))
                month = int(input("Tháng (1-12): "))
                year = int(input("Năm: "))
                if 1 <= day <= 31 and 1 <= month <= 12 and year > 0:
                    self.ngay_vao_cong_ty = Date(day, month, year)
                    break
                print("Ngày hoặc tháng không hợp lệ. Vui lòng nhập lại.")
            except ValueError:
                print("Vui lòng nhập số nguyên hợp lệ.")
    
    def display(self):
        print(f"Họ tên: {self.ho_ten}")
        print("Ngày sinh:", end=" ")
        self.ngay_sinh.display()
        print("Ngày vào công ty:", end=" ")
        self.ngay_vao_cong_ty.display()

# Chương trình quản lý nhân viên
n = int(input("Nhập số lượng nhân viên: "))
danh_sach_nv = []

for i in range(n):
    print(f"\nNhập thông tin nhân viên thứ {i+1}:")
    nv = Employee()
    nv.nhap_thong_tin()
    danh_sach_nv.append(nv)

print("\nDanh sách nhân viên:")
for i, nv in enumerate(danh_sach_nv, 1):
    print(f"Nhân viên {i}:")
    nv.display()
    print("---")
```

```python
# bai 9:
import math

class DaGiac:
    def __init__(self):
        pass
    
    def nhap_thong_tin(self):
        pass
    
    def tinh_chu_vi(self):
        return 0
    
    def tinh_dien_tich(self):
        return 0
    
    def in_thong_tin(self):
        print("Chu vi:", self.tinh_chu_vi())
        print("Diện tích:", self.tinh_dien_tich())

class HinhBinhHanh(DaGiac):
    def __init__(self):
        self.canh_a = 0
        self.canh_b = 0
        self.goc = 0  # Góc giữa hai cạnh, đơn vị độ
    
    def nhap_thong_tin(self):
        self.canh_a = float(input("Nhập cạnh a của hình bình hành: "))
        self.canh_b = float(input("Nhập cạnh b của hình bình hành: "))
        self.goc = float(input("Nhập góc giữa hai cạnh (độ): "))
    
    def tinh_chu_vi(self):
        return 2 * (self.canh_a + self.canh_b)
    
    def tinh_dien_tich(self):
        return self.canh_a * self.canh_b * math.sin(math.radians(self.goc))

class HinhChuNhat(HinhBinhHanh):
    def __init__(self):
        super().__init__()
        self.goc = 90  # Góc cố định 90 độ
    
    def nhap_thong_tin(self):
        self.canh_a = float(input("Nhập chiều dài hình chữ nhật: "))
        self.canh_b = float(input("Nhập chiều rộng hình chữ nhật: "))
    
    def tinh_dien_tich(self):
        return self.canh_a * self.canh_b

class HinhVuong(HinhChuNhat):
    def __init__(self):
        super().__init__()
    
    def nhap_thong_tin(self):
        self.canh_a = float(input("Nhập cạnh hình vuông: "))
        self.canh_b = self.canh_a
    
    def tinh_chu_vi(self):
        return 4 * self.canh_a
    
    def tinh_dien_tich(self):
        return self.canh_a ** 2

# Chương trình chính để demo
print("Nhập thông tin hình bình hành:")
hbh = HinhBinhHanh()
hbh.nhap_thong_tin()
print("Thông tin hình bình hành:")
hbh.in_thong_tin()

print("\nNhập thông tin hình chữ nhật:")
hcn = HinhChuNhat()
hcn.nhap_thong_tin()
print("Thông tin hình chữ nhật:")
hcn.in_thong_tin()

print("\nNhập thông tin hình vuông:")
hv = HinhVuong()
hv.nhap_thong_tin()
print("Thông tin hình vuông:")
hv.in_thong_tin()
```

```python
# bai 10:
import math

class Diem:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
    
    def nhap_thong_tin(self):
        self.x = float(input("Nhập tọa độ x: "))
        self.y = float(input("Nhập tọa độ y: "))

class Clip(Diem):
    def __init__(self, x=0, y=0, x2=0, y2=0):
        super().__init__(x, y)
        self.x2 = x2
        self.y2 = y2
    
    def nhap_thong_tin(self):
        super().nhap_thong_tin()
        self.x2 = float(input("Nhập tọa độ x2 của điểm cuối: "))
        self.y2 = float(input("Nhập tọa độ y2 của điểm cuối: "))
    
    def do_dai(self):
        dx = self.x2 - self.x
        dy = self.y2 - self.y
        return math.sqrt(dx * dx + dy * dy)

class DuongTron(Clip):
    def __init__(self, x=0, y=0, x2=0, y2=0):
        super().__init__(x, y, x2, y2)
    
    def nhap_thong_tin(self):
        super().nhap_thong_tin()
        while True:
            ban_kinh = self.do_dai() / 2
            if ban_kinh > 0:
                break
            print("Điểm đầu và điểm cuối không hợp lệ để tạo đường tròn. Vui lòng nhập lại.")
            super().nhap_thong_tin()
    
    def tinh_dien_tich(self):
        ban_kinh = self.do_dai() / 2
        return math.pi * ban_kinh * ban_kinh
    
    def in_thong_tin(self):
        ban_kinh = self.do_dai() / 2
        print(f"Bán kính: {ban_kinh}")
        print(f"Diện tích: {self.tinh_dien_tich()}")

# Chương trình chính
dt = DuongTron()
print("Nhập thông tin đường tròn (dùng hai điểm để xác định bán kính):")
dt.nhap_thong_tin()
print("\nThông tin đường tròn:")
dt.in_thong_tin()
```
